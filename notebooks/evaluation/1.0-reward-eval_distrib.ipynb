```python
import numpy as np
import plotly.graph_objects as go

def plot_3d_reward_histogram(df, column='pos_rewards', k=5):
    data_for_plot = []
    steps = df['global_step'].unique()

    for step in steps[::-1][::k]:
        step_data = df[df['global_step'] == step][column].dropna()  # Exclude NaN values
        if not step_data.empty:  # Exclude if no data
            hist, bin_edges = np.histogram(step_data, bins=70)
            xs = (bin_edges[:-1] + bin_edges[1:]) / 2  # Compute center of bins
            data_for_plot.append(go.Scatter3d(x=xs, y=[step]*len(xs), z=hist, mode='lines'))

    # Create the 3D plot
    fig = go.Figure(data=data_for_plot)
    fig.update_layout(scene=dict(xaxis_title=column,
                                 yaxis_title='Global Step',
                                 zaxis_title='Count'))
    fig.show(fullscreen=True)
```
